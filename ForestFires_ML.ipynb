### Stratified sampling

```python
%matplotlib inline
import matplotlib.pyplot as plt
ForestFires.hist(column='area', bins=10)
plt.show()
```

```python
import numpy as np
ForestFires["area_after"] = np.ceil(ForestFires["area"]*(2/3))
ForestFires["area_after"].where(ForestFires["area"] < 8, 8.0, inplace=True)
ForestFires["area_after"].hist()
plt.show()
```

```python
X = ForestFires.drop("area", axis = 1)
Y = ForestFires["area"]
train_X, test_X, train_Y, test_Y = train_test_split(X,Y, test_size=0.2, random_state=42, stratify=ForestFires['area_after'])
```

```python
train_X
```

```python
train_X = train_X.drop("area_after", axis=1)
test_X = test_X.drop("area_after", axis=1)

train_X.reset_index(drop=True, inplace=True)
test_X.reset_index(drop=True, inplace=True)
```

```python
train_X
```

```python
test_X
```

### Handle categorical attributes

```python
ForestFires_cat1 = train_X["month"]
ForestFires_cat_encoded, ForestFires_categories = ForestFires_cat1.factorize()
print('ForestFires_categories: ',ForestFires_categories)
```

```python
ForestFires_cat2 = train_X["day"]
ForestFires_cat_encoded2, ForestFires_categories2 = ForestFires_cat2.factorize()
print('ForestFires_categories2: ',ForestFires_categories2)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder_month = OneHotEncoder()
ForestFires_cat1hot = encoder_month.fit_transform(ForestFires_cat_encoded.reshape(-1,1))
ForestFires_cat1hot
```

```python
encoder_day = OneHotEncoder()
ForestFires_cat2hot = encoder_day.fit_transform(ForestFires_cat_encoded2.reshape(-1,1))
ForestFires_cat2hot
```

```python
ForestFires_cat1hot=ForestFires_cat1hot.toarray()
ForestFires_cat2hot=ForestFires_cat2hot.toarray()
```

```python
ForestFires_cat_month2 = pd.DataFrame(ForestFires_cat1hot,columns=ForestFires_categories.tolist())
ForestFires_cat_day2 = pd.DataFrame(ForestFires_cat2hot,columns=ForestFires_categories2.tolist())
ForestFires_cat = pd.concat([ForestFires_cat_month2, ForestFires_cat_day2], axis=1)
ForestFires_cat
```

```python
train_X.drop(["month","day"], inplace = True, axis =1)
train_X.reset_index(drop=True, inplace=True)
train_X
```


### Handle outliers

```python
Q1 = train_X.quantile(0.25)
Q3 = train_X.quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
#all the rows that has outliers will be True
outlier_index = ((train_X < (Q1 - 1.5 * IQR)) |(train_X > (Q3 + 1.5 * IQR))).any(axis=1)
outlier_index.reset_index(drop=True, inplace=True)

print(outlier_index.value_counts())
```

### Data transformation and scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(train_X)
scaled=scaler.transform(train_X)
ForestFires_scaled = pd.DataFrame(scaled, columns=train_X.columns)
ForestFires_scaled
```

```python
#reset index of dataframes before merge
ForestFires_cat.reset_index(drop=True, inplace=True)
ForestFires_scaled.reset_index(drop=True, inplace=True)

#merge 
train_X = pd.concat([ForestFires_scaled, ForestFires_cat], axis=1)
train_X
```

### Repeat the steps to be implemented on the test set

```python
#handle categorical attributes
ForestFires_cat3 = test_X["month"]
ForestFires_cat_encoded3, ForestFires_categories3 = ForestFires_cat3.factorize()
```

```python
ForestFires_cat4 = test_X["day"]
ForestFires_cat_encoded4, ForestFires_categories4 = ForestFires_cat4.factorize()
```

```python
ForestFires_cat3hot = encoder_month.fit_transform(ForestFires_cat_encoded3.reshape(-1,1))
ForestFires_cat4hot = encoder_day.fit_transform(ForestFires_cat_encoded4.reshape(-1,1))
```

```python
ForestFires_cat3hot=ForestFires_cat3hot.toarray()
ForestFires_cat4hot=ForestFires_cat4hot.toarray()
```

```python
ForestFires_cat_month3 = pd.DataFrame(ForestFires_cat3hot, columns=ForestFires_categories3.tolist())
ForestFires_cat_day4 = pd.DataFrame(ForestFires_cat4hot, columns=ForestFires_categories4.tolist())
```

```python
ForestFires_cat2 = pd.concat([ForestFires_cat_month3, ForestFires_cat_day4], axis=1)
ForestFires_cat2
```

```python
test_X.drop(["month","day"], inplace = True, axis =1)
test_X.reset_index(drop=True, inplace=True)
test_X
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(test_X)
scaled = scaler.transform(test_X)
ForestFires_scaled2 = pd.DataFrame(scaled, columns=test_X.columns)

#reset index of dataframes before merge
ForestFires_cat2.reset_index(drop=True, inplace=True)
ForestFires_scaled2.reset_index(drop=True, inplace=True)

#merge 
test_X = pd.concat([ForestFires_scaled2, ForestFires_cat2], axis=1)
test_X 
```

### Feature selection

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE
forest_reg = RandomForestRegressor(n_estimators=100 )

#select the best 8 features
rfe_ForestFires = RFE(estimator=forest_reg, n_features_to_select=8, step=1)
rfe_ForestFires.fit(train_X, train_Y.values.ravel())   
reduced_ForestFires=rfe_ForestFires.transform(train_X)
```

```python
forest_reg.fit(reduced_ForestFires, train_Y.values.ravel())
```

### Random Forest Regression Cross Validation

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())


from sklearn.model_selection import cross_validate
forest_output = cross_validate(forest_reg, reduced_ForestFires, train_Y.values.ravel(), scoring="neg_mean_squared_error", cv=10,  return_estimator =True)
forest_rmse_scores = np.sqrt(-forest_output["test_score"])
display_scores(forest_rmse_scores)
```

```python
cols = list(train_X.columns) 
temp = pd.Series(rfe_ForestFires.support_,index = cols) 
selected_features_rfe = temp[temp==True].index 
print(selected_features_rfe)
```

```python
test_X = test_X[selected_features_rfe]
test_X
```

```python
train_X = train_X[selected_features_rfe]
train_X
```

 Part 2:  Models Training

 Linear Regression


```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(train_X, train_Y.values.ravel())
train_Y.reset_index(drop=True, inplace=True)

some_data = train_X.iloc[:5]
some_labels = train_Y.iloc[:5]
print("Predictions:", lin_reg.predict(some_data))
print("Labels:", some_labels)
```

### Measure the performance 

```python
#calculate distance
from sklearn.metrics import mean_squared_error
import numpy as np
ForestFires_predictions = lin_reg.predict(train_X)
lin_mse = mean_squared_error(train_Y, ForestFires_predictions)
lin_rmse = np.sqrt(lin_mse)

lin_rmse 
```

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor(random_state=42)

tree_reg.fit(train_X, train_Y)
ForestFires_predictions = tree_reg.predict(train_X)
tree_mse = mean_squared_error(train_Y, ForestFires_predictions)
tree_rmse = np.sqrt(tree_mse)
tree_rmse 
```

### Decision Tree Reg Cross Validation

```python
from sklearn.model_selection import cross_validate
output = cross_validate(tree_reg, train_X, train_Y, scoring="neg_mean_squared_error", cv=10, return_estimator =True)
tree_rmse_scores = np.sqrt(-output["test_score"])

display_scores(tree_rmse_scores) 
```

### Linear Reg Cross Validation

```python
lin_output = cross_validate(lin_reg, train_X, train_Y.values.ravel(), scoring="neg_mean_squared_error", cv=10, return_estimator =True)
lin_rmse_scores = np.sqrt(-lin_output["test_score"])

display_scores(lin_rmse_scores)
```

### Random Forest Regressor Cross Validation

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_validate

forest_reg = RandomForestRegressor(n_estimators=10 )
forest_reg.fit(train_X, train_Y.values.ravel())

forest_output = cross_validate(forest_reg, train_X, train_Y.values.ravel(), scoring="neg_mean_squared_error", cv=10,  return_estimator =True)
forest_rmse_scores = np.sqrt(-forest_output["test_score"])

display_scores(forest_rmse_scores)
```
